```python
import pandas as pd

events4['new_date'] = pd.to_datetime(
    events4['DATE'],
    errors='coerce' 
)
events4['new_date'] = events4['new_date'].ffill()

events4['new_date'] = events4['new_date'].apply(
    lambda x: x.strftime("%Y-%m-%d %H:%M:%S") if pd.notnull(x) else pd.NA
)

print(events4)
```

```python
index_9 = events4[events4['N°'] == 9].index.tolist()
print(f"\nIndex avec N°=9: {index_9}")

# Si on a au moins 2 index avec 9, renommer le deuxième
if len(index_9) >= 2:
    deuxieme_index = index_9[1]
    events4.loc[deuxieme_index, 'N°'] = 10
print(events4)
```

```python
def corriger_biat(valeur):
    if pd.isna(valeur):
        return valeur
    
    valeur_str = str(valeur).strip()
    
   
    if valeur_str.upper() == 'BIAT':
        return 'BIAT €'
    
   
    return valeur_str


events4['Banque'] = events4['Banque'].apply(corriger_biat)
```
