```python
import pandas as pd
local2022= pd.read_excel("DATA FORMATION BI 2022.xlsx",index_col=False,keep_default_na=True,sheet_name='LOCAL', header=4)
print(local2022.head())
```

```python
msno.bar(local2022)
plt.show()   
```

```python
missing_values = local2022.isnull().sum()
missing_percent = (missing_values / len(local2022)) * 100
missing_table = pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_percent})
print(missing_table) 
```

```python
local2022['NOM DU CLIENT'] = local2022['NOM DU CLIENT'].fillna('pas de description')
```

```python
local2022['N°'] = range(1, 30) 
```
